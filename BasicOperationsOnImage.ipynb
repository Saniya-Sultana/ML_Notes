```python
features = []
class_labels = []
for folder in folders:
    for images in os.listdir(r"C:\Users\91912\Downloads\flower_photos\{}".format(folder)):
        img = cv2.imread(r"C:\Users\91912\Downloads\flower_photos\{}\{}".format(folder,images),0)
        img = cv2.resize(img,(50,50)) # resize the image
        img = img.flatten()  # Falttening array
        features.append(img)  
        class_labels.append(folder)
```

#### Task - Showing a image in b,g,r shades

```python
img = cv2.imread(r"C:\Users\91912\Downloads\flower_photos\tulips\38287568_627de6ca20.jpg")
```

```python
img
```

```python
img.shape
```

```python
b,g,r = cv2.split(img)
```

```python
zeroes = np.zeros((333,500),dtype = np.uint8)
```

```python
blue_channel = cv2.merge([b,zeroes,zeroes])
green_channel = cv2.merge([zeroes,g,zeroes])
red_channel = cv2.merge([zeroes,zeroes,r])
```

```python
cv2.imshow("Blue_Channel",blue_channel)
cv2.imshow("Green_Channel",green_channel)
cv2.imshow("Red_Channel",red_channel)
cv2.imshow("Orginal_img",cv2.merge([b,g,r]))

cv2.waitKey(0)

cv2.destroyAllWindows()
```
