## Color Images

#### Red color image

```python
b = np.zeros((500,500),dtype=np.uint8)
g = np.zeros((500,500),dtype=np.uint8)
r = np.full((500,500),255,dtype=np.uint8)

color_img1=cv2.merge([b,g,r])   ### Merging the three channels into 3d array
```

#### Blue color image

```python
b = np.full((500,500),255,dtype=np.uint8)
g = np.zeros((500,500),dtype=np.uint8)
r = np.zeros((500,500),dtype=np.uint8)

color_img2=cv2.merge([b,g,r])
```

#### Green color image

```python
b = np.zeros((500,500),dtype=np.uint8)
g = np.full((500,500),255,dtype=np.uint8)
r = np.zeros((500,500),dtype=np.uint8)

color_img3=cv2.merge([b,g,r])
```

#### Displaying the different color arrays into color image formats

```python
cv2.imshow("red",color_img1)
cv2.imshow("blue",color_img2)
cv2.imshow("green",color_img3)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
